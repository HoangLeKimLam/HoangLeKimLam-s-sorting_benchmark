```python


import numpy as np
import pandas as pd

# Hằng số cấu hình
NUM_SAMPLES = 1_000_000
LOW_BOUND = -10**6
HIGH_BOUND = 10**6

# Đặt seed để tái lập kết quả (reproducibility)
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

print("Khởi tạo hoàn tất.")
```

```python

#DÃY TĂNG DẦN (INTEGER & FLOAT)


#  Integer tăng dần
ascending_int = np.sort(
    np.random.randint(LOW_BOUND, HIGH_BOUND + 1, NUM_SAMPLES)
)

#  Float tăng dần
ascending_float = np.sort(
    np.random.uniform(LOW_BOUND, HIGH_BOUND, NUM_SAMPLES)
)

print("Tạo dãy tăng dần hoàn tất.")
```

```python

# DÃY GIẢM DẦN (INTEGER & FLOAT)


# Integer giảm dần
descending_int = np.sort(
    np.random.randint(LOW_BOUND, HIGH_BOUND + 1, NUM_SAMPLES)
)[::-1]

# Float giảm dần
descending_float = np.sort(
    np.random.uniform(LOW_BOUND, HIGH_BOUND, NUM_SAMPLES)
)[::-1]

print("Tạo dãy giảm dần hoàn tất.")
```

```python

#  4 DÃY INTEGER NGẪU NHIÊN


random_int_1 = np.random.randint(LOW_BOUND, HIGH_BOUND + 1, NUM_SAMPLES)
random_int_2 = np.random.randint(LOW_BOUND, HIGH_BOUND + 1, NUM_SAMPLES)
random_int_3 = np.random.randint(LOW_BOUND, HIGH_BOUND + 1, NUM_SAMPLES)
random_int_4 = np.random.randint(LOW_BOUND, HIGH_BOUND + 1, NUM_SAMPLES)

print("Tạo 4 dãy integer ngẫu nhiên hoàn tất.")
```

```python

# 4 DÃY FLOAT NGẪU NHIÊN


random_float_1 = np.random.uniform(LOW_BOUND, HIGH_BOUND, NUM_SAMPLES)
random_float_2 = np.random.uniform(LOW_BOUND, HIGH_BOUND, NUM_SAMPLES)
random_float_3 = np.random.uniform(LOW_BOUND, HIGH_BOUND, NUM_SAMPLES)
random_float_4 = np.random.uniform(LOW_BOUND, HIGH_BOUND, NUM_SAMPLES)

print("Tạo 4 dãy float ngẫu nhiên hoàn tất.")
```

```python

#  TỔNG HỢP DATAFRAME


dataset = pd.DataFrame({
    
    # ---- INTEGER ----
    "INT_ASCENDING": ascending_int,
    "INT_DESCENDING": descending_int,
    "INT_RANDOM_1": random_int_1,
    "INT_RANDOM_2": random_int_2,
    "INT_RANDOM_3": random_int_3,
    "INT_RANDOM_4": random_int_4,
    
    # ---- FLOAT ----
    "FLOAT_ASCENDING": ascending_float,
    "FLOAT_DESCENDING": descending_float,
    "FLOAT_RANDOM_1": random_float_1,
    "FLOAT_RANDOM_2": random_float_2,
    "FLOAT_RANDOM_3": random_float_3,
    "FLOAT_RANDOM_4": random_float_4,
})

print("Tạo DataFrame hoàn tất.")
print(dataset.shape)
```

```python
print(dataset)
```

```python

OUTPUT_FILENAME = "synthetic_sort_data.csv"

# Xuất file
dataset.to_csv(
    OUTPUT_FILENAME,
    index=False,          # không ghi chỉ số dòng
    float_format="%.6f"   # định dạng float cho đẹp và gọn
)

print(f"Đã lưu file thành công: {OUTPUT_FILENAME}")
```
